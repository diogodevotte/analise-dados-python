```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ler os dados
base_dados = pd.read_csv('unicorns till sep 2022.csv')

# Dimensão
base_dados.shape
```

```python
# Primeiros registros
base_dados.head()
```

```python
base_dados.columns
```

```python
#Renomear
base_dados.rename( columns={
    'Company' : 'Empresas',
    'Valuation ($B)' : 'Valor ($)',
    'Date Joined' : 'Data de adesão',
    'Country' : 'País',
    'City\xa0' : 'Cidade',
    'Industry' :  'Setor',
    'Investors' : 'Investidores'
}, inplace=True ) # Mudanças na origem
```

```python
display(base_dados)
```

```python
# Verificar o tipo da informação
base_dados.info()
```

```python
# Verificar a quantidade de campos nulos
base_dados.isnull().sum()
```

```python
# Gráfico de campos nulos
plt.figure( figsize=(15,6))
plt.title('Analisando os Campos Nulos')
sns.heatmap(base_dados.isnull(), cbar=False)
```

```python
# Visualizando os Setores
plt.figure( figsize=(15,6))
plt.title('Análise dos Setores')
plt.bar(base_dados['Setor'].value_counts().index,base_dados['Setor'].value_counts())
plt.xticks( rotation=45 , ha='right');
```

```python
# Percentual de unicórnios por país
analise = round(base_dados['País'].value_counts( normalize = True) * 100,1)
```

```python
analise
```

```python
# Plot geral dos países
plt.figure( figsize=(15,6))
plt.title('Análise dos Países Geradores de Unicórnio')
plt.pie(
    analise,
    labels = analise.index,
    shadow = True,
    startangle = 90,
    autopct = '%1.1f%%'
);
```

```python
# Plot top 10 países
plt.figure( figsize=(15,6))
plt.title('Análise dos Países Geradores de Unicórnios - Top 10')
plt.pie(
    analise.head(10),
    labels = analise.index[0:10],
    shadow = True,
    startangle = 90,
    autopct = '%1.1f%%'
);
```

```python
# Conversão para Data
base_dados['Data de adesão'] = pd.to_datetime(base_dados['Data de adesão'])
base_dados['Data de adesão'].head()
```

```python
# Extrair o Ano e Mês
base_dados['Mês'] = pd.DatetimeIndex(base_dados['Data de adesão']).month
base_dados['Ano'] = pd.DatetimeIndex(base_dados['Data de adesão']).year
display(base_dados)
```

```python
# Tabela analítica - agrupando os dados
analise_agrup = base_dados.groupby( by=['País','Ano','Mês','Empresas']).count()['Data de adesão'].reset_index()
```

```python
# Filtrando o país
analise_agrup.loc[
    analise_agrup['País'] == 'Brazil'
]
```

```python
# apply: percorrer linha por linha e realizar operações
# base_dados['Valor ($)'] = pd.to_numeric( base_dados['Valor ($)'].apply(lambda Linha: Linha.replace('$','')))
base_dados['Valor ($)'].head()
```

```python
# Analisando os países
analise_pais = base_dados.groupby( by=['País'])['Valor ($)'].sum().reset_index()
analise_pais.head()
```

```python
# Ordernar em ordem crescente
analise_valor = analise_pais.sort_values('Valor ($)', ascending=False)
analise_valor.head()
```

```python
# Países em relação ao valor gerado pelos unicórnios
plt.figure( figsize=(15,6))
plt.title('Países em Relação ao Valor dos Unicórnios')
plt.xticks( rotation=45 , ha='right')
plt.plot( analise_valor['País'], analise_valor['Valor ($)'])
```
