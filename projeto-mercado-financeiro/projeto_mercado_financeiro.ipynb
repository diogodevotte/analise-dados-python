```python
# Libs para modelagem e matrizes
import numpy as np
import pandas as pd

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
# Lendo e verificando os dados
base_dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')
print(base_dados.shape)
base_dados.head()
```

```python
# Info
base_dados.info()
```

```python
base_dados.describe()
```

```python
# Séries Temporais
dados = base_dados.set_index('Data')
dados.head()
```

```python
# Verificar estilos disponíveis
print(plt.style.available)
```

```python
# Analisando as ações da Magalu
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15,6))
plt.title('Análise das ações da Magalu - Fechamento', fontsize=18, loc='left')
plt.plot(dados.index, dados['Fechamento'])
plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
# Verifica os 5 últimos valores da base de dados
dados.tail()
```

```python
# Média móvel
media_movel = dados['Fechamento'].rolling(5).mean()
media_tendencia = dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(15,6))
plt.title('Análise das ações da Magalu - Fechamento', fontsize=18, loc='left')
plt.plot(dados.index, dados['Fechamento'])
plt.plot(dados.index, media_movel, color='red')
plt.plot(dados.index, media_tendencia, color = 'green')
plt.xlabel('Período da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
sns.boxplot( data=dados, x='Fechamento')
```

```python
# Boxplot Mensal
base_dados['Mês'] = base_dados['Data'].dt.month
plt.figure(figsize=(16,5))
sns.boxplot(data=base_dados,x='Mês',y='Fechamento')
```

```python
# Análise analítica
base_dados.groupby(by='Mês').describe()['Fechamento']
```

```python
# Gráfico de ações
grafico = go.Figure(
    data = [
        go.Candlestick(
            x = dados.index,
            open = dados['Abertura'],
            high = dados['Maior'],
            low = dados['Menor'],
            close = dados['Fechamento'],
        )
    ]
)

grafico.show()
```
